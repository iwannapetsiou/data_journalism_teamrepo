## ΠΕΡΙΛΗΨΗ 
<i> Στο πρότζεκτ μας θέλουμε να δούμε αν η άνοδος της θερμοκρασίας ορισμένων χωρών συσχετίζεται με την άνοδο των ακραίων καιρικών φαινομένων. Οι χώρες που μελετώνται είναι η Ελλάδα,  η Ιταλία, η Πορτογαλία και η Ισπανία, καθώς πρόκεται για χώρες που έχουν παρόμοια γεωγραφικά, κλιματικά και γεωλογικά χαρακτηριστικά. Η σύγκριση γίνεται καθώς θέλουμε να δούμε αν ό,τι ισχύει για την Ελλάδα ισχύει και για άλλες παρόμοιες γεωλογικά χώρες, ή αν οφείλεται σε άλλους παράγοντες.<i/>

<i> Το διάστημα που θα μελετηθεί είναι από το 1997 έως το 2022. Για να μελετηθεί η αύξηση της θερμοκρασίας και η αύξηση των ακραίων καιρικών φαινομένων χρειάζεται ένα ακόμη χρονικό διάστημα που θα λειτουργήσει ως σύγκριση για το 1997-2022. Το διάστημα αυτό είναι από το 1972 έως το 1996. Συμπερασματικά, επιδιώκουμε να αναδείξουμε πως η θερμοκρασία και η συχνότητα ακραίων καιρικών φαινομέων έχουν αυξηθεί το διάστημα 1997-2022 σε σύγκριση με το 1972-1996.</i>

<i> Τα ακραία καιρικά φαινόμενα που θα μελετηθούν είναι: φωτιές, πλημμύρες, ακραίες θερμοκρασίες και καταιγίδες. </i>
